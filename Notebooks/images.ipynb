```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define CNN layer info for the formatted diagram
layers_info = [
    ("Input\n(64643)", 'input'),
    ("Conv2D (16)\n(646416)", 'conv'),
    ("MaxPool\n(323216)", 'pool'),
    ("Conv2D (32)\n(323232)", 'conv'),
    ("MaxPool\n(161632)", 'pool'),
    ("Conv2D (64)\n(161664)", 'conv'),
    ("MaxPool\n(8864)", 'pool'),
    ("Conv2D (128)\n(88128)", 'conv'),
    ("MaxPool\n(44128)", 'pool'),
    ("Flatten\n(2048)", 'flat'),
    ("Dense (512)", 'dense'),
    ("Dropout (0.2)", 'dropout'),
    ("Dense (256)", 'dense'),
    ("Dropout (0.2)", 'dropout'),
    ("Dense (7)\nSoftmax", 'output')
]

# Color scheme by type
layer_colors = {
    'input': '#aed581',
    'conv': '#64b5f6',
    'pool': '#81d4fa',
    'flat': '#ffd54f',
    'dense': '#ce93d8',
    'dropout': '#ffe082',
    'output': '#e57373'
}

# Initialize plot
fig, ax = plt.subplots(figsize=(20, 4.5))
ax.axis('off')

# Draw each layer box
for i, (label, ltype) in enumerate(layers_info):
    x = 0.02 + i * 0.065
    y = 0.4
    width = 0.06
    height = 0.2

    # Draw rounded box
    box = patches.FancyBboxPatch(
        (x, y), width, height,
        boxstyle="round,pad=0.02",
        linewidth=1.5,
        edgecolor='black',
        facecolor=layer_colors.get(ltype, '#ffffff')
    )
    ax.add_patch(box)
    ax.text(x + width / 2, y + height / 2, label, ha='center', va='center', fontsize=8)

    # Draw arrow to next layer
    if i < len(layers_info) - 1:
        ax.annotate('', xy=(x + width, y + height / 2), xytext=(x + width + 0.005, y + height / 2),
                    arrowprops=dict(arrowstyle='->', lw=1.3))

plt.title("Custom CNN Architecture with Output Shapes", fontsize=14)
plt.show()
```
